```python
import csv
import pandas as pd
import numpy as np

spamReader = csv.reader(open('../City_Budget_Expenditures.csv', newline=''), delimiter=' ', quotechar='|')
#for row in spamReader:
  #  print(', '.join(row))
df = pd.read_csv("City_Budget_Expenditures.csv")
df.head()
df["BUDGET CHANGE"]= df["BUDGET CHANGE AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()

df["TOTAL EXPENDITURES"]= df["TOTAL EXPENDITURES"].replace('[\$,]','',regex=True).astype('double')

df["ADOPTED BUDGET AMOUNT"]= df["ADOPTED BUDGET AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()

df["BUDGET TRANSFER IN"]= df["BUDGET TRANSFER IN AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()

df["BUDGET TRANSFER OUT AMOUNT"]= df["BUDGET TRANSFER OUT AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()

df["TOTAL BUDGET"]= df["TOTAL BUDGET"].replace('[\$,]','',regex=True).astype('double')

df["ENCUMBRANCE AMOUNT"]= df["ENCUMBRANCE AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()

df["PRE-ENCUMBRANCE AMOUNT"]= df["PRE-ENCUMBRANCE AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()

df["BUDGET UNCOMMITTED AMOUNT"]= df["BUDGET UNCOMMITTED AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()



df['DIFFERENCE BETWEEN BUDGET AND EXPENDITURES'] = df['TOTAL BUDGET'] - df['TOTAL EXPENDITURES']

df["BUDGET CHANGE AVERAGE"]= df["BUDGET CHANGE AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()

df["TOTAL EXPENDITURES AVERAGE"]= df["TOTAL EXPENDITURES"].replace('[\$,]','',regex=True).astype('double').mean()

df["ADOPTED BUDGET AVERAGE"]= df["ADOPTED BUDGET AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()

df["BUDGET TRANSFER IN AVERAGE"]= df["BUDGET TRANSFER IN AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()

df["BUDGET TRANSFER OUT AVERAGE"]= df["BUDGET TRANSFER OUT AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()

df["TOTAL BUDGET AVERAGE"]= df["TOTAL BUDGET"].replace('[\$,]','',regex=True).astype('double').mean()

df["ENCUMBRANCE AVERAGE"]= df["ENCUMBRANCE AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()

df["PRE-ENCUMBRANCE AVERAGE"]= df["PRE-ENCUMBRANCE AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()

df["BUDGET UNCOMMITTED AVERAGE"]= df["BUDGET UNCOMMITTED AMOUNT"].replace('[\$,]','',regex=True).astype('double').mean()

df['TOTAL BUDGET AVERAGE'] = df["TOTAL BUDGET"].replace('[\$,]','',regex=True).astype('double').mean()
df['TOTAL EXPENDITURE AVERAGE'] = df["TOTAL BUDGET"].replace('[\$,]','',regex=True).astype('double').mean()
df.groupby(['FUND NAME','DEPARTMENT NAME'])['TOTAL BUDGET', 'DIFFERENCE BETWEEN BUDGET AND EXPENDITURES','TOTAL EXPENDITURES','TOTAL BUDGET AVERAGE','TOTAL EXPENDITURE AVERAGE','BUDGET CHANGE AVERAGE','TOTAL EXPENDITURES AVERAGE','ADOPTED BUDGET AVERAGE','BUDGET TRANSFER IN AVERAGE','BUDGET TRANSFER OUT AVERAGE','ENCUMBRANCE AVERAGE','PRE-ENCUMBRANCE AVERAGE','BUDGET UNCOMMITTED AVERAGE'].mean() 

```

Data Story:
Based upon some quick data analysis and visualizations, one interesting case that stands out to me is the breakdown of how the total budget counts have changed since 2012. Seeing how the city has changed its spending behaviors and patterns, specific in which industries, helps to understand the ways in which the city is changing its focus to help improve Los Angeles as a whole. The department that consistently receives the most funding, around 220-240 expenses every year, is Transportation. While most other departments average between about 30-70, the Transportation department by far accounts for the largets portion of city expenditures over the years. This is interesting because it reveals many realities about Los Angeles as a city. LA is very much reliant upon transporation, and clearly the city expenditures reveal that the city government is fully devoted to ensuring there are practical measures being funded in order to build up city more beneficial to varying forms of transit. With programs like expanding the METRO and working to lessen the dependence on cars and other forms of unsustainable private transportation, the goverment is devoted to increasing the role of public transportation within the city of LA and the budget expenditures starting from 2012 provide more clarity to this story. Another interesting piece of data is many firms are running a defecit within the year. Calculating the difference between total budgets and total expenditures, the difference is often negative. This trend could be explained by firms focusing on improving their programs and increasing their investment in order to reap larger rewards and benefits later in the future. Firms are comfortable with creating a spending gap because programs like improving the METRO and other public policies offer the opportunity for increased rewards and benefits in the future.
